# 7. 辞書を作成

```python
tagger = MeCab.Tagger()
tagger.parse(txt)
```

```python
tagger.parse(txt).split("\n") #改行で分ける
```

```python
elements = tagger.parse(txt).split("\n")[:-2] #後ろ2つは不要な要素なので削除
element = elements[0]
element
```

```python
parts = element.split(",")
parts
```

```python
surface_pos, pos1, base = parts[0], parts[1], parts[-3] #表層形_品詞, 品詞1(品詞の種類？), 原型
surface, pos = surface_pos.split("\t") #始めまして\t感動詞を分割

# 分割できたので、forで辞書を作成する
```

```python
results = []

for element in elements:
    parts = element.split(",")
    surface_pos, pos1, base = parts[0], parts[1], parts[-3]
    surface, pos = surface_pos.split("\t")
    results.append(dict(表層形=surface, 基本形=base, 品詞=pos, 品詞1=pos1)) #dictクラスを使用

results
```

## 8. ユニークな動詞の表層形・基本形の抽出

```python
surface_verbs = set()

for result in results:
    if result["品詞"] == "動詞":
        surface_verbs.add(result["表層形"])

surface_verbs
```

```python
base_verbs = set()

for result in results:
    if result["品詞"] == "動詞":
        base_verbs.add(result["基本形"])

base_verbs
```

## 9. 単語の出現頻度の算出

```python
from collections import defaultdict

word_freq =  defaultdict(int)

for result in results:
    if result["品詞"] != "記号":
        word_freq[result["基本形"]] += 1
        # キーの宣言と同時に値を入れているイメージ(多分)

word_freq
```

## 14. n-gram化する関数の作成

```python
def ngram(text, n):
    return [text[i:i+n] for i in range(len(text)-n+1)]
```

## 19. k-meansによるクラスタリング

```python
max_vocab = 1000 # 1000単語もってくる
vocab = model.index_to_key[:max_vocab] # 1000の語彙リストを作成
vectors = [model[word] for word in vocab] # 単語ベクトルリストを作成

display(vocab[1])
display(vectors[1])
```

```python
from sklearn.cluster import KMeans
import numpy as np

n_clusters = 100 # 1000個の単語を100個に分類する
vectors_np = np.array(vectors) # vectorsをnumpy配列に変換

kmeans_model = KMeans(n_clusters=n_clusters, verbose=1, random_state=42)
# k-meansのインスタンスをここで作成する
# verboseは進行状況をコンソールに表示するかどうか指定する
# random_stateは乱数の取り方を数字で指定するもので、42がよく使用される

kmeans_model.fit(vectors_np) 
```

```python
cluster_labels = kmeans_model.labels_
# ラベルを配列に格納
# ラベルは、いわば分類の種名
# vocab = ["うどん", "そば", "ラーメン"]なら、cluster_labels = [0, 0, 1]みたいなイメージ

cluster_to_words = defaultdict(list)
for cluster_id, word in zip(cluster_labels, vocab):
    cluster_to_words[cluster_id].append(word)
# cluster_to_words = {0:"うどん", "そば", 1:"ラーメン"}みたいな辞書を作る
```

```python
for words in cluster_to_words.values():
  print(words[:10])
# 100個の分類の頭10個をそれぞれ表示
```

## 20. t-SNEによる可視化
t-SNEを用いると、単語ベクトルの次元を削減し、2次元上に単語ベクトルを可視化することができる

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=123) # 2次元に次元削減
embedded = tsne.fit_transform(np.array(vectors).astype(np.float64))

display(embedded)
```

```python
plt.figure(figsize=(20, 20))
plt.scatter(np.array(embedded).T[0], np.array(embedded).T[1]) # x座標とy座標をそれぞれとってくる
for (x, y), name in zip(embedded, vocab):
    plt.annotate(name, (x, y))
plt.show()
```
